```python
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib.ticker import ScalarFormatter

df_dqn = pd.read_csv('csv/ddqn.csv')
df_q_learning = pd.read_csv('csv/q_learning.csv')

# plt.figure(figsize=(10, 5))  # Set the figure size
plt.ticklabel_format(axis='x', style='sci', scilimits=(0, 6))
plt.xlim(0, 10 * 10**5)
plt.plot(df_dqn['Iteration'], df_dqn['Avg Throughput'], label='DQN')
plt.plot(df_q_learning['Iteration'], df_q_learning['Avg Throughput'], label='Q_learning')
plt.ylabel('Avg reward', fontsize = 13)
plt.xlabel('Iterations', fontsize = 13)
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.grid(True)
plt.tight_layout()
plt.legend()
formatter = ScalarFormatter(useMathText=True)
plt.gca().xaxis.set_major_formatter(formatter)
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df_htt_throughput = pd.read_csv('csv/htt_throughput.csv')
df_dqn_throughput = pd.read_csv('csv/ddqn_throughput.csv')

plt.plot(df_htt_throughput['d_t'], df_htt_throughput['throughput'], marker='o', label='HTT')
plt.plot(df_dqn_throughput['d_t'], df_dqn_throughput['throughput'], marker='x', label='DQN')
plt.xlabel('dt')
plt.ylabel('Avg throughput (packages/time unit)')
plt.title('HTT vs Deep Q-Learning Avarage Throughput')
plt.tight_layout()
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df_htt_loss = pd.read_csv('csv/htt_loss.csv')
df_ddqn_loss = pd.read_csv('csv/ddqn_loss.csv')

plt.plot(df_htt_loss['d_t'], df_htt_loss['package_loss'], marker='o', label='HTT')
plt.plot(df_ddqn_loss['d_t'], df_ddqn_loss['package_loss'], marker='x', label='DQN')
plt.xlabel('dt')
plt.ylabel('Avg loss (packages/time unit)')
plt.title('HTT vs Deep Q-Learning Packages Loss')
plt.tight_layout()
plt.legend()
plt.show()
```
